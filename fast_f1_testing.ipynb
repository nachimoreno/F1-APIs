```python
YEAR = 2026
```

```python
schedule = f1.get_event_schedule(YEAR)
schedule.head(3) # schedule object
```

```python
aus = schedule.get_event_by_round(1)
aus # event object
```

```python
aus_p2 = aus.get_practice(2)
aus_p2.load()
aus_p2
```

```python
aus_p2_laps = aus_p2.laps
aus_p2_laps.sample(10)
```

```python
aus_p2_laps.drop(columns=['LapStartDate', 'Position', 'Sector1SessionTime', 'Sector2SessionTime', 'Sector3SessionTime', 'LapStartTime'], inplace=True)
```

```python
aus_p2_laps.dropna(subset=["LapTime"], inplace=True)
aus_p2_laps = aus_p2_laps[
    ~(
        aus_p2_laps['PitOutTime'].notna() |
        aus_p2_laps['PitInTime'].notna()
    )
]
```

```python
aus_p2_laps[
    (aus_p2_laps['Driver'] == 'PIA') & (aus_p2_laps['Stint'] == 1)
]
```

```python
cooldown_threshold_seconds = 4  # <-- adjust this
cooldown_threshold = pd.to_timedelta(cooldown_threshold_seconds, unit='s')
```

```python
# Filter out cooldown laps (laps > X seconds slower than stint fastest)
aus_p2_laps_filtered = (
    aus_p2_laps
    .assign(
        FastestLapInStint=lambda dataframe:
            dataframe.groupby(['Driver', 'Stint'])['LapTime'].transform('min')
    )
)

aus_p2_laps_filtered = aus_p2_laps_filtered[
    aus_p2_laps_filtered['LapTime'] <=
    aus_p2_laps_filtered['FastestLapInStint'] + cooldown_threshold
]

# 1) Aggregate with descriptive names (avoids MultiIndex columns)
aus_p2_summary = (
    aus_p2_laps_filtered
    .groupby(['Driver', 'Stint'])
    .agg(
        LapCount=('Stint', 'count'),
        FastestLap=('LapTime', 'min'),
        StintPace=('LapTime', 'mean')
    )
    .reset_index()
)

# 2) Compute driver order by their overall fastest lap (smallest timedelta = fastest)
driver_fastest = (
    aus_p2_summary
    .groupby('Driver')['FastestLap']
    .min()
    .sort_values(ascending=True)
)

# 3) Turn Driver into a categorical with the desired order so sorting keeps drivers grouped
aus_p2_summary['Driver'] = pd.Categorical(
    aus_p2_summary['Driver'],
    categories=driver_fastest.index.tolist(),
    ordered=True
)

# 4) Sort by Driver then by StintPace (fastest → slowest within driver)
aus_p2_summary = aus_p2_summary.sort_values(
    by=['Driver', 'StintPace'],
    ascending=[True, True]
).reset_index(drop=True)

# 5) Optional: readable formatted columns
def format_timedelta_series_to_m_ss_mmm(td_series, milliseconds=3):
    total_seconds = td_series.dt.total_seconds()
    minutes = (total_seconds // 60).astype(int)
    seconds = total_seconds % 60
    fmt = minutes.astype(str) + ":" + seconds.map(lambda s: f"{s:06.{milliseconds}f}")
    return fmt

aus_p2_summary['_FastestLap'] = format_timedelta_series_to_m_ss_mmm(aus_p2_summary['FastestLap'])
aus_p2_summary['_StintPace'] = format_timedelta_series_to_m_ss_mmm(aus_p2_summary['StintPace'])

# 6) MultiIndex view
aus_p2_summary_indexed = aus_p2_summary.set_index(['Driver', 'Stint'])

aus_p2_summary_indexed.drop(columns=['FastestLap', 'StintPace'])[aus_p2_summary_indexed['LapCount'] > 1]
```
